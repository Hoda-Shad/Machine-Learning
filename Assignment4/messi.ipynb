```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Dataset/Lionel Messi Goals.csv')
```

```python
data["Date"] = pd.to_datetime(data["Date"], format='%Y%m%d')
data['Date'] = data['Date'].map(lambda date: date.year )
```

```python
data.head(10)
```

```python
goalperyear = data['Score Home Team - When Messi Scored'].groupby(data['Date']).sum().reset_index()
```

```python
goalperyear
```

```python
plt.bar(goalperyear['Date'], goalperyear['Score Home Team - When Messi Scored'])
plt.title("Messi Goals per year")
plt.show()
```
